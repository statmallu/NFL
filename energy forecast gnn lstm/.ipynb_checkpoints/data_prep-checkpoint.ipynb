## Train-Test Splitting for LSTM with Non-Overlapping Windows
Problem with Typical Train-Test Splits
In many machine learning tasks, especially time series forecasting, it's common to split the dataset into training and testing sets. However, when using rolling windows for forecasting, there's a risk of data leakage. This happens when part of the testing data overlaps with the training data, leading to unrealistically low errors during evaluation. This situation violates the assumption of complete independence between training and testing datasets, which is crucial for building robust models.

Proposed Solution
To address this, we use a discrete train-test splitting strategy where we:
1) Partition the dataset by months to account for seasonal dependencies.
2) Randomly sample dates within each month without replacement to create distinct training and testing sets.
    a) A percentage (e.g., 70%) of dates is selected for the training set.
    b) The remaining dates are used for the testing set.
3) This ensures:
    a) Both the training and testing sets cover data across the entire year, preserving seasonality patterns.
    b) There is no overlap between training and testing, avoiding data leakage.

### How Random Sampling Works
1) Within each month, the method shuffles the available dates.
2) It then splits these dates into two sets:
    a) Training dates: Randomly sampled train_frac (e.g., 70%) of dates from each month.
    b) Testing dates: The remaining dates for that month.
Because sampling is random and without replacement, the method does not always pick certain days consistentlyit selects dates randomly. This ensures variability and prevents any specific pattern from being hardcoded into the split.

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import numpy as np
np.random.seed(42)

# Parameters
train_frac = 0.7  # Fraction of dates to use for training
n_days = 7  # Number of past days to use
time_steps = 24 * n_days  # Total time steps for past data
forecast_steps = 24  # Forecast next day
import numpy as np
import pandas as pd

continuous_vars = ['car1', 'grid', 'solar', 'temperature', 'dew_point', 'humidity',
                   'visibility', 'apparent_temperature', 'pressure', 'wind_speed',
                   'cloud_cover', 'wind_bearing']
target_vars = ['grid', 'solar']
categorical_cols = ['hour', 'day_of_week', 'month']

# --- Preprocessing: One-Hot Encoding ---
# Instantiate encoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform categorical variables
encoded_cats = encoder.fit_transform(merged_data[categorical_cols])

# Create a DataFrame for encoded features
encoded_cats_df = pd.DataFrame(
    encoded_cats,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=merged_data.index
)

# Merge encoded features back into the dataset
merged_data = pd.concat([merged_data, encoded_cats_df], axis=1)

# Drop original categorical columns
merged_data = merged_data.drop(columns=categorical_cols)

# Update the list of categorical columns to use in LSTM windows
categorical_cols = encoder.get_feature_names_out(categorical_cols).tolist()

# --- Preprocessing: MinMax Scaling ---
scaler = MinMaxScaler()

# Fit and transform continuous variables
scaled_continuous = scaler.fit_transform(merged_data[continuous_vars])

# Replace scaled data back in the DataFrame
scaled_continuous_df = pd.DataFrame(
    scaled_continuous,
    columns=continuous_vars,
    index=merged_data.index
)
merged_data.update(scaled_continuous_df)

# --- Train/Test Split ---
# Group data by month
merged_data['month'] = merged_data['localhour'].dt.month
monthly_groups = merged_data.groupby('month')

# Split each month's dates into train and test
train_dates, test_dates = [], []
for month, group in monthly_groups:
    unique_dates = group['localhour'].dt.date.unique()
    np.random.shuffle(unique_dates)
    
    split_idx = int(len(unique_dates) * train_frac)
    train_dates.extend(unique_dates[:split_idx])
    test_dates.extend(unique_dates[split_idx:])

# Assign train/test labels to the data
merged_data['set'] = np.where(merged_data['localhour'].dt.date.isin(train_dates), 'train', 'test')

# Split into train and test data
train_data = merged_data[merged_data['set'] == 'train']
test_data = merged_data[merged_data['set'] == 'test']

# Drop the 'set' column as it's no longer needed
train_data = train_data.drop(columns=['set'])
test_data = test_data.drop(columns=['set'])
```

```python
def create_lstm_windows(data, time_steps, forecast_steps, continuous_vars, target_vars, categorical_cols):
    X_past, X_future_weather, Y = [], [], []
    for idx in range(len(data) - time_steps - forecast_steps):
        # Past data
        past_data = data.iloc[idx : idx + time_steps]
        
        # Continuous features for the past
        X_continuous = past_data[continuous_vars].values
        
        # Categorical features for the past
        X_categorical = past_data[categorical_cols].values
        
        # Future weather features
        future_targets = data.iloc[idx + time_steps : idx + time_steps + forecast_steps]
        X_weather_future = future_targets[['temperature', 'dew_point', 'humidity', 'visibility',
                                           'apparent_temperature', 'pressure', 'wind_speed', 
                                           'cloud_cover', 'wind_bearing']].values
        
        # Targets
        Y_targets = future_targets[target_vars].values
        
        # Append data to lists
        X_past.append(np.hstack([X_continuous, X_categorical]))  # Combine past continuous and categorical
        X_future_weather.append(X_weather_future)  # Append future weather separately
        Y.append(Y_targets)
    
    return np.array(X_past), np.array(X_future_weather), np.array(Y)
```

```python

# Create rolling windows for training and testing datasets
X_train_past, X_train_weather_future, Y_train = create_lstm_windows(
    data=train_data,
    time_steps=time_steps,
    forecast_steps=forecast_steps,
    continuous_vars=continuous_vars,
    target_vars=target_vars,
    categorical_cols=encoded_cats_df.columns  # Use encoded categorical columns
)

X_test_past, X_test_weather_future, Y_test = create_lstm_windows(
    data=test_data,
    time_steps=time_steps,
    forecast_steps=forecast_steps,
    continuous_vars=continuous_vars,
    target_vars=target_vars,
    categorical_cols=encoded_cats_df.columns  # Use encoded categorical columns
)

# Check shapes
print("Train Past Input shape (X_train_past):", X_train_past.shape)  # [num_train_samples, time_steps, num_features]
print("Train Future Weather shape (X_train_weather_future):", X_train_weather_future.shape)  # [num_train_samples, forecast_steps, num_weather_features]
print("Train Output shape (Y_train):", Y_train.shape)  # [num_train_samples, forecast_steps, num_targets]
print("Test Past Input shape (X_test_past):", X_test_past.shape)
print("Test Future Weather shape (X_test_weather_future):", X_test_weather_future.shape)
print("Test Output shape (Y_test):", Y_test.shape)
```

```python
import os
os.makedirs("./data/lstm", exist_ok=True)

# Save data
np.savez_compressed("./data/lstm/train_data.npz", 
                    X_train_past=X_train_past, 
                    X_train_weather_future=X_train_weather_future, 
                    Y_train=Y_train)
np.savez_compressed("./data/lstm/test_data.npz", 
                    X_test_past=X_test_past, 
                    X_test_weather_future=X_test_weather_future, 
                    Y_test=Y_test)
```

## Data preparation for Graph Neural Network (GNN)

    1) Nodes: Each node represents a feature at a given time step (e.g., weather, energy consumption, or categorical variables).
    2) Edges: Relationships between features (e.g., correlation or similarity) or sequential connections for temporal relationships.
    3) Prediction Goal: Predict the energy consumption using learned relationships between feature nodes over time.

### Step1. Define Nodes
    1) Each node is a (feature, time) pair. For example:
        Features: ['car1', 'grid', 'solar', 'temperature', 'hour', 'day_of_week']
        Time Steps: Last 7 days (7  24 = 168)
        The total number of nodes = n_features * 168.
### Step2. Define Edges
        Temporal Edges: Connect a feature at time t to the same feature at t+1.
        Cross-Feature Edges: Connect features at the same time step based on correlation or similarity.
### Step3. Node Features
        The node features will include:
        Value of the Feature: At the specific time step.
        Static Information: Encoded into the feature node (e.g., household information).
### Step4. Prediction Mechanism
        Since energy consumption is not a direct node attribute:
            Aggregate Embeddings: Aggregate the learned embeddings for all nodes at a time step.
            Decoder: Use an MLP or another mechanism to map the aggregated embedding to a single prediction (e.g., energy consumption for that time step).

### Graph Construction


```python
# Node preparation
import numpy as np
import pandas as pd
import networkx as nx
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import numpy as np

# Define variables
continuous_vars = [
    'car1', 'grid', 'solar', 'temperature', 'dew_point', 'humidity',
    'visibility', 'apparent_temperature', 'pressure', 'wind_speed',
    'cloud_cover', 'wind_bearing'
]
categorical_cols = ['hour', 'day_of_week', 'month']

# Initialize preprocessors
scaler = MinMaxScaler()
encoder = OneHotEncoder(sparse_output=False)

# Scale continuous variables
scaled_continuous = scaler.fit_transform(merged_data[continuous_vars])
print(f"Scaled continuous variables shape: {scaled_continuous.shape}")  # Expect: (n_samples, len(continuous_vars))

# One-hot encode categorical variables
encoded_categorical = encoder.fit_transform(merged_data[categorical_cols])
print(f"Encoded categorical variables shape: {encoded_categorical.shape}")  # Expect: (n_samples, sum(unique values in categorical_cols))

# Combine features
features = np.hstack([scaled_continuous, encoded_categorical])
print(f"Combined features shape: {features.shape}")  # Expect: (n_samples, len(continuous_vars) + sum(unique values in categorical_cols))

# Extract target variable (grid)
target_data = merged_data[['grid']].values
print(f"Target data shape: {target_data.shape}")  # Expect: (n_samples, 1)
```

```python
# Node-level preparation
n_days = 7  # Use past 7 days
n_hours = 24  # Hours per day
n_steps = n_days * n_hours  # Total number of hours

household_ids = merged_data['dataid'].unique()
node_features = []
node_targets = []

for household in household_ids:
    # Extract data for the household
    household_data = merged_data[merged_data['dataid'] == household]
    
    # Ensure sufficient data is available
    if len(household_data) < n_steps + 24:
        continue  # Skip households with insufficient data
    
    # Prepare node features: Use past `n_steps` hours
    recent_data = household_data.iloc[-(n_steps + 24):-24]
    household_features = features[recent_data.index]
    node_features.append(household_features.mean(axis=0))  # Aggregate features
    
    # Prepare node targets: Use next 24 hours for grid
    future_targets = target_data[recent_data.index][-24:]
    node_targets.append(future_targets.flatten())

node_features = np.array(node_features)
node_targets = np.array(node_targets)

print(f"Node features shape: {node_features.shape}")  # Expect: (n_nodes, len(features[0]))
print(f"Node targets shape: {node_targets.shape}")  # Expect: (n_nodes, 24)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

# Step 1: Compute cosine similarity between node features
similarity_matrix = cosine_similarity(node_features)
print(f"Similarity matrix shape: {similarity_matrix.shape}")  # Expect: (n_nodes, n_nodes)

# Step 2: Define edges based on a similarity threshold or top-k neighbors
threshold = 0.8  # Similarity threshold for edge creation
adjacency_matrix = (similarity_matrix > threshold).astype(int)  # Binary adjacency matrix
print(f"Adjacency matrix shape: {adjacency_matrix.shape}")  # Expect: (n_nodes, n_nodes)

# Optional: Retain top-k similar neighbors for each node
k = 5
adjacency_matrix_topk = np.zeros_like(similarity_matrix)
for i in range(similarity_matrix.shape[0]):
    top_k_indices = similarity_matrix[i].argsort()[-k:]  # Indices of top-k similar nodes
    adjacency_matrix_topk[i, top_k_indices] = similarity_matrix[i, top_k_indices]

# Step 3: Create a graph using NetworkX
G = nx.from_numpy_array(adjacency_matrix_topk)
print(f"Graph has {G.number_of_nodes()} nodes and {G.number_of_edges()} edges.")

# Step 4: Add edge weights based on similarity
for i, j in G.edges():
    G[i][j]['weight'] = similarity_matrix[i, j]

# Step 5: Convert edge list and weights into tensors
edge_index = np.array(list(G.edges)).T  # Convert to edge index format (2, num_edges)
edge_weights = np.array([G[i][j]['weight'] for i, j in G.edges])  # Extract edge weights

print(f"Edge index shape: {edge_index.shape}")  # Expect: (2, num_edges)
print(f"Edge weights shape: {edge_weights.shape}")  # Expect: (num_edges,)
```

```python
# Define file path
file_path = './data/gnn/graph_data.npz'

# Save node features, targets, edge index, and edge weights
np.savez(
    file_path,
    node_features=node_features,
    node_targets=node_targets,
    edge_index=edge_index,
    edge_weights=edge_weights
)

print(f"Data saved to {file_path}")
```
